```python
# Merge two dictionaries 

capitals_dict = {**capitals_dict, **the_rest}
capitals_dict
```

```python
'''
This was a difficult one for me! Note the structure of the code.

The following code randomly selects a state from a list and asks the user to input the capital of a US state. 
If the user gets it right, it prints 'Correct!'.
If the user doesn't get it right, it repeatedly asks the user for an input until either they get it right 
at which point it prints 'Correct!' or, type 'exit' at which point, it prints the correct name of the 
capital city and goodbye.

Things to note:
1) continue breaks out of the while loop and goes back to the line that starts the while loop
2) break skips the code contained in the while loop code block and goes to the line after the while loop code block
3) print(a var) prints 'a var'; a var prints a var
4) .title() capitalizes the first letter of each word in a phrase
5) .capitalize() capitalizes all letters in a word
'''

import random

state = random.choice(list(capitals_dict.keys()))
capital = capitals_dict[state].lower()
user_input = input(f'Enter the capital of {state}: ').lower()
correct_answer = False

while correct_answer is False:
    user_input = input(f'Enter the capital of {state}: ')
    
    if user_input == capital:
        correct_answer = True
    else:
        correct_answer = False

    
    if user_input == 'exit':
        print(f'{capital.title()} Goodbye!')
        break
    if correct_answer == True:
        print('Correct!')
        break
```
